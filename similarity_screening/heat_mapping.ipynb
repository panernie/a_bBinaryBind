```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

# AA_DATA

```python
df=pd.read_excel("data/AA_DATA.xlsx")
```

```python
df.head(1)
```

```python
df.set_index('name', inplace=True)
```

```python
print(df.loc[:,['Berberine']])
```

```python
#plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Arial'] + plt.rcParams['font.serif']
#plt.rcParams['font.weight'] = 'bold'  # 设置字体为加粗
plt.figure(figsize=(20, 15))
sns.set(font_scale=1.4)
heatmap = sns.heatmap(df.loc[:,['Berberine']],square=True,annot=True,cbar=True,annot_kws={"fontsize":16},cbar_kws={"shrink": 0.5})
plt.xticks(fontsize=18)  
plt.yticks(fontsize=18) 
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=24) 
#plt.ylabel('Similarity score', fontsize=34, fontname='Times New Roman', fontweight='bold')
#plt.savefig('AA_heatmap.png', dpi=300) 
plt.show()
```

# DFT Bind

```python
df=pd.read_excel("data/bind.xlsx")
df.set_index('name', inplace=True)
```

```python
df
```

```python
plt.rcParams['font.serif'] = ['Arial'] + plt.rcParams['font.serif']
#plt.rcParams['font.weight'] = 'bold'  # 设置字体为加粗

plt.figure(figsize=(15, 18))
sns.set(font_scale=2.5)
heatmap = sns.heatmap(df,square=True,annot=True,cbar=True,fmt=".3f", annot_kws={"weight": "bold","fontsize":24},cbar_kws={"shrink": 0.7})

# 调整热图中的标签字体大小
plt.xticks(fontsize=24,fontname='Arial', fontweight='bold',rotation=20)  # 设置x轴标签字体大小为14
plt.yticks(fontsize=24,fontname='Arial', fontweight='bold',rotation=0)  # 设置y轴标签字体大小为14


# 调整图例（colorbar）的字体大小
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=24)  # 设置图例的字体大小为14

#plt.xlabel('X-axis Label')
plt.ylabel('Binding energy (kcal/mol)', fontsize=42, fontname='Arial', fontweight='bold')
#plt.savefig('Binding energy.png', dpi=300) 
plt.show()
```

```python
df=pd.read_excel("data/total.xlsx")
df.set_index('name', inplace=True)
```

```python
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
#plt.rcParams['font.weight'] = 'bold'  # 设置字体为加粗

plt.figure(figsize=(15, 18))
sns.set(font_scale=2.5)
heatmap = sns.heatmap(df,square=True,annot=True,cbar=True,fmt=".3f", annot_kws={"weight": "bold","fontsize":24},cbar_kws={"shrink": 0.4})

# 调整热图中的标签字体大小
plt.xticks(fontsize=24,fontname='Times New Roman', fontweight='bold',rotation=20)  # 设置x轴标签字体大小为14
plt.yticks(fontsize=24,fontname='Times New Roman', fontweight='bold',rotation=0)  # 设置y轴标签字体大小为14

# 调整图例（colorbar）的字体大小
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=24)  # 设置图例的字体大小为14
plt.subplots_adjust(top=1.2)
#plt.xlabel('X-axis Label')
plt.ylabel('Total energy (Eh)', fontsize=42, fontname='Times New Roman', fontweight='bold')
#plt.subplots_adjust(right=0.1)
#plt.savefig('Total_energy.png', dpi=300) 
plt.show()
```
